```python
# WEEK 5 RELATED PRACTICE
#Python range() with float values-Instead of range(), arange will be used as range() is not applicable for float values
# Quick examples of getting range of float values
import pandas as pd
import numpy as np
```

```python
# np.arange() without step parameter- by default step parameter is 1 until a number less that stop value(5.5)
# arange() - syntax : arange([start,] stop[, step,], dtype=None, *, like=None)
#start parameter specifies the starting value
#stop parameter specifies the end value of the range
#step will increment the value within the specified range.by default it is 1
#Method 1
for i in np.arange(0.5,5.5):# This condition of i works as long as i is with in the range of (0.5<=i<5.5)- excluding th stop value
    # end=', ' is meant to avoid printing of each at new lline. Hence this allows us to print all on the same line 
    print(i, end=', ') 
```

```python
#https://sparkbyexamples.com/python-tutorial/python-range-with-float-values/
# np.arange() with step parameter
#Method 2
for i in np.arange(0.5,5.5,1.5):
     print(i, end=', ')
```

```python
# Using np.linspace()- # Syntax : numpy.linspace(start,stop,num,endpoint)
#The linspace() function in numpy is also used to create the sequences of evenly spaced values within a defined interval.
#start parameter specifies the starting value
#stop parameter specifies the end value of the range
#num will generate a total number of specified within the range of values. By default, it is 50.
#endpoint is used to stop the last sample. By default it is True.

#Method 3
for i in np.linspace(1.5, 5.5):#by default num is 50. that is 50 points are generated if num is not specified
    print(i, end=', ') 
```

```python
# or Method 3
y = [x for x in np.linspace(1.5, 5.5)]#by default num is 50. that is 50 points are generated if num is not specified
y
```

```python
df2 = pd.DataFrame(y)
df2
```

```python
# Using np.linspace() with num parameter
#Method 4
for i in np.linspace(1.5, 5.5,num=5):# num = 5 which is 5 points in the range, it includes the stop value  
    print(i, end=', ') # to print out all on horizontal line that is end = ' ' avoids newline
```

```python
import pandas as pd
technologies = {
    'Courses':["Spark","PySpark","Python","pandas"],
    'Fee' :[20000,25000,22000,30000],
    'Duration':['30days','40days','35days','50days'],
              }
index_labels=['r1','r2','r3','r4']
df1 = pd.DataFrame(technologies,index=index_labels)
df1
```

```python
technologies2 = {
    'Courses':["Spark","Java","Python","Go"],
    'Discount':[2000,2300,1200,2000]
              }
index_labels2=['r1','r6','r3','r5']
df2 = pd.DataFrame(technologies2,index=index_labels2)
df2
```

```python
# Using pandas.merge()
df3= pd.merge(df1,df2) # by default it is merging on the common values  that is inner - displays what is common
df3   
```

```python
#Merge DataFrames on Columns 
df4=pd.merge(df1,df2, on='Courses')# which is the same as the one mentioned above
df4
```

```python
# When column names are different
df5=pd.merge(df1,df2, left_on='Courses', right_on='Courses')
print(df5)
```

```python
# Merge by left Join
df6=pd.merge(df1,df2, on='Courses', how='left')
print(df6)
```

```python
# Merge by right Join
df6=pd.merge(df1,df2, on='Courses', how='right')
print(df6)
```

```python
# Merge by outer Join
df7=pd.merge(df1,df2, on='Courses', how='outer')
print(df7)
```

```python

import pandas as pd
technologies   = ({
    'Courses':["Spark","PySpark","Hadoop","Python","Pandas","Hadoop","Spark","Python","NA"],
    'Fee' :[22000,25000,23000,24000,26000,25000,25000,22000,1500],
    'Duration':['30days','50days','55days','40days','60days','35days','30days','50days','40days'],
    'Discount':[1000,2300,1000,1200,2500,None,1400,1600,0]
          })
df = pd.DataFrame(technologies)
print(df)
```

```python
#It is used for grouping the data points (i.e. rows) based on the distinct values in the given column or columns.
#We can then calculate aggregated values for the generated groups.
# Use groupby() to compute the sum
# groupby ([ "  column label  "]) gives grouped distinct vales with their corresponding other columns
df2 =df.groupby(['Courses']).sum()
print(df2)
```

```python
# Use groupby() to compute the sum
df2 =df.groupby(['Courses']).mean() # it is finding mean of each distinct values. ex hadoop has frequency of two with 23000 and 25000. mean of the two will be (23k+25k)/2 = 48k/2 = 24k
# that is fee and amount are all mean values
print(df2)
```
